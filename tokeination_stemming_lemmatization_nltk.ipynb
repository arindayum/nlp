```python
import nltk
from nltk.tokenize import word_tokenize, TweetTokenizer, MWETokenizer, TreebankWordTokenizer
from nltk.stem import PorterStemmer, SnowballStemmer, WordNetLemmatizer
import string

# Download required datasets (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "Hello! This is a sample sentence. Let's test tokenization, stemming, and lemmatization :) #NLP"
```

```python
#whitespace tokenisation
whitespace_tokens = text.split()
print("Whitespace Tokenization:", whitespace_tokens)
```

```python
#punctuation based
punctuation_tokens = []
for word in whitespace_tokens:
    word = word.strip(string.punctuation)
    if word:
        punctuation_tokens.append(word)

print("Punctuation-based Tokenization:", punctuation_tokens)
```

```python
# trebank tokenisation
treebank = TreebankWordTokenizer()
treebank_tokens = treebank.tokenize(text)
print("Treebank Tokenization:", treebank_tokens)
```

```python
#tweet tokenisation
tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(text)
print("Tweet Tokenization:", tweet_tokens)
```

```python
#mwe tokenisation
mwe_tokenizer = MWETokenizer([('sample', 'sentence'), ('test', 'tokenization')])
mwe_tokens = mwe_tokenizer.tokenize(word_tokenize(text))
print("MWE Tokenization:", mwe_tokens)
```

```python
#porten stemmer
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in punctuation_tokens]
print("Porter Stemming:", porter_stems)
```

```python
#snowball stemmer
snowball = SnowballStemmer("english")
snowball_stems = [snowball.stem(word) for word in punctuation_tokens]
print("Snowball Stemming:", snowball_stems)
```

```python
#lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in punctuation_tokens]
print("Lemmatization:", lemmatized_words)
```

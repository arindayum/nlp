```python
# ================================
# INSTALL (Run once if needed)
# ================================
# !pip install nltk scikit-learn gensim

import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import numpy as np

# Download tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# ================================
# DATASET
# ================================
corpus = [
    "This is a sample sentence",
    "This sentence is for testing",
    "Word embeddings are useful in NLP"
]
```

```python
# ================================
# 1. BAG OF WORDS (COUNT)
# ================================
print("\n--- BAG OF WORDS (COUNT) ---")

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

vocab = vectorizer.get_feature_names_out()
bow_matrix = X.toarray()

print("Vocabulary:", vocab)
print("BoW Matrix:\n", bow_matrix)
```

```python
# ================================
# 2. NORMALIZED COUNT (TERM FREQUENCY)
# ================================
print("\n--- NORMALIZED COUNT ---")

normalized_bow = bow_matrix / bow_matrix.sum(axis=1, keepdims=True)

print("Normalized BoW:\n", normalized_bow)
```

```python
# ================================
# 3. TF-IDF
# ================================
print("\n--- TF-IDF ---")

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus)

print("Vocabulary:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
# ================================
# 4. WORD2VEC EMBEDDINGS
# ================================
print("\n--- WORD2VEC ---")

# Tokenize sentences
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# Train Word2Vec model
model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=3,
    min_count=1,
    workers=4
)

# Example word vector
print("\nVector for 'sentence':")
print(model.wv['sentence'])

# Similar words
print("\nWords similar to 'sentence':")
print(model.wv.most_similar('sentence'))
```
